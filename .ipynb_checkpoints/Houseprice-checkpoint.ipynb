## Correlation

```python
numeric_features = data.select_dtypes(include=[np.number])
numeric_features.columns 
```

```python
categorical_features = data.select_dtypes(include = [object])
categorical_features.columns
```

```python
correlation = numeric_features.corr()
print(correlation['SalePrice'].sort_values(ascending=False),'\n')
```

```python
f, ax = plt.subplots(figsize =(14 ,12))
plt.title('Correlation of Numeric Features with Sale Price ',y=1,size=16)
sns.heatmap(correlation,square = True, vmax = .8)
```

```python
k = 11
cols = correlation.nlargest(k , 'SalePrice')['SalePrice'].index
print(cols)
cm = np.corrcoef(data[cols].values.T)
f, ax = plt.subplots(figsize = (14,12))
sns.heatmap(cm, vmax =.8, linewidth = 0.01, square = True, annot = True, cmap='viridis',linecolor ="white", xticklabels = cols.values, annot_kws = {'size':12},yticklabels = cols.values)
```

```python
first_quartile = data['SalePrice'].quantile(.25)
third_quartile = data['SalePrice'].quantile(.75)
IQR = third_quartile-first_quartile
```

```python
new_boundary = third_quartile + 3*IQR
```

```python
data.drop(data[data['SalePrice']>new_boundary].index, axis = 0, inplace = True)
```
